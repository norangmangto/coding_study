# Advent of Code 2025

## Day 1

### part1

```python
def solution_day1():
    count_zero = 0
    curr_number = 50

    with open("inputs/day1.txt") as f:
        for line in f:
            direction = line[0]
            steps = int(line[1:])

            if direction == "L":
                curr_number -= steps
                curr_number += 100
            elif direction == "R":
                curr_number += steps
                curr_number -= 100

            curr_number = curr_number % 100

            if curr_number == 0:
                count_zero += 1

    return count_zero

solution_day1()
```

### Part 2

```python
def solution_day1_part2():
    count_zero = 0
    curr_number = 50

    with open("inputs/day1.txt") as f:
        for line in f:
            direction = line[0]
            steps = int(line[1:])
            count_zero += steps//100
            steps %= 100

            if direction == "L":
                new_number = curr_number - steps
                if curr_number != 0 and new_number <= 0:
                    new_number += 100
                    count_zero += 1
            elif direction == "R":
                new_number = curr_number + steps
                if new_number >= 100:
                    new_number -= 100
                    count_zero += 1

            curr_number = new_number % 100

    return count_zero

solution_day1_part2()
```

## Day 2

### Part 1

```python
def solution_day2_part1():
    added_up_invalid_ids = 0
    invalid_ids = []

    with open("inputs/day2.txt") as f:
        raw_id_ranges = f.read().split(",")

    for id_range in raw_id_ranges:
        id_range = id_range.strip()
        start_id, end_id = map(int, id_range.split("-"))

        for id in range(start_id, end_id + 1):
            # the ID is invalid if it is made only of some sequence of digits repeated twice.
            id_str = str(id)
            length = len(id_str)
            if length % 2 == 0:
                half = length // 2
                if id_str[:half] == id_str[half:]:
                    invalid_ids.append(id)
                    added_up_invalid_ids += id
                    continue

    return added_up_invalid_ids, invalid_ids

solution_day2_part1()
```

### Part 2

```python
def solution_day2_part2():
    added_up_invalid_ids = 0
    invalid_ids = []

    with open("inputs/day2.txt") as f:
        raw_id_ranges = f.read().split(",")

    for id_range in raw_id_ranges:
        id_range = id_range.strip()
        start_id, end_id = map(int, id_range.split("-"))

        for id in range(start_id, end_id + 1):
            # an ID is invalid if it is made only of some sequence of digits repeated at least twice.
            id_str = str(id)
            length = len(id_str)
            for sub_length in range(1, length // 2 + 1):
                if length % sub_length == 0:
                    times = length // sub_length
                    sub_str = id_str[:sub_length]
                    if sub_str * times == id_str:
                        invalid_ids.append(id)
                        added_up_invalid_ids += id
                        break

    return added_up_invalid_ids, invalid_ids

solution_day2_part2()
```

## Day 3

### Part 1

```python
def solution_day3_part1():
    total_joltage = 0

    with open("inputs/day3.txt") as f:
        for line in f:
            line = line.strip()

            first_val = second_val = "-1"
            first_idx = -1

            # Find the first largest value and its index
            for i, v in enumerate(line[:-1]):
                if int(v) > int(first_val):
                    first_val = v
                    first_idx = i

            for i, v in enumerate(line[first_idx + 1:]):
                if int(v) > int(second_val):
                    second_val = v

            joltage = int(first_val + second_val)
            print(f"Line: {line}, Joltage: {joltage}")

            total_joltage += joltage

    return total_joltage

solution_day3_part1()
```

### Part 2

```python
def solution_day3_part2():
    BATTERY_COUNT = 12
    total_joltage = 0

    with open("inputs/day3.txt") as f:
        for line in f:
            line = line.strip()
            l = len(line)

            joltage = 0
            prev_idx = -1

            for c in range(BATTERY_COUNT, 0, -1):
                max_val = "-1"
                max_val_idx = -1
                print(f"{c=}, {prev_idx=}, {line[prev_idx+1:l-c+1]}")

                for i, v in enumerate(line[prev_idx+1:l-c+1]):
                    if int(v) > int(max_val):
                        max_val = v
                        max_val_idx = prev_idx + 1 + i

                joltage += int(max_val) * pow(10, c-1)
                prev_idx = max_val_idx
                print(f"c: {c}, max_val: {max_val}, max_val_idx: {max_val_idx}, joltage: {joltage}, prev_idx: {prev_idx}")

            print(f"Line: {line}, Joltage: {joltage}")
            total_joltage += joltage

    return total_joltage

solution_day3_part2()
```
